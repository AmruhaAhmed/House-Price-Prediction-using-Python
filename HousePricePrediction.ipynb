```python
def mae(max_leaf_nodes,X_train,X_test,Y_train,Y_test):
  drt2=DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=1)
  drt2.fit(X_train,Y_train)
  val_pred2=drt.predict(X_test)
  val_mae2=mean_absolute_error(val_pred2,Y_test)
  return val_mae2
```

```python
max_leaf_nodes=[20,30,40,50,60,70,80,90,100,150,200,10,500,300,250,400,450,600,700]
chnode=20000000
minmae=10000
for i in max_leaf_nodes:
  maeobtained=mae(i,X_train,X_test,Y_train,Y_test)
  print("\n {} is mae and {} is decision tree".format(maeobtained,i))
  if(maeobtained<minmae):
    chnode=i
    minmae=maeobtained



best_tree_size=chnode
```

```python
drt3=DecisionTreeRegressor(max_leaf_nodes=best_tree_size,random_state=1)
drt3.fit(X_train,Y_train)
valuepredict=(drt3.predict(X_test))
```

```python
print(mean_absolute_error(valuepredict,Y_test))
```
